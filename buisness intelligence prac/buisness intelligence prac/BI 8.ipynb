```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have already performed the ETL process and loaded the data into a DataFrame

# Sample DataFrame
data = {
    'Year': [2019, 2020, 2021, 2022, 2023],
    'Revenue': [100000, 120000, 150000, 180000, 200000],
    'Expenses': [60000, 65000, 70000, 75000, 80000]
}

df = pd.DataFrame(data)

# Visualize revenue and expenses over the years
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='Revenue', label='Revenue')
sns.lineplot(data=df, x='Year', y='Expenses', label='Expenses')
plt.title('Revenue and Expenses Over the Years')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.legend()
plt.grid(True)
plt.show()
```
